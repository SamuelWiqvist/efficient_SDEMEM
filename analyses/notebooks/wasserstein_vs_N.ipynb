Load sigma_eps for kalman 

```python
R = 60000
burn_in = 10000

N_time = "200"
M_subjects = "40"

for i in range(len(seeds)):
    seed = str(seeds[i])

    chain___kalman = np.loadtxt(open("data/SDEMEM OU/kalman for plot mess vs N/chain_sigma_epsilon_" + seed + "_" + M_subjects + "_" + N_time + ".csv", "rb"), delimiter=",", skiprows=1)
    chain___kalman = chain___kalman[(burn_in+1):R]

    # select last 5k samples
    chain___kalman = chain___kalman[-n:]
    data_kalman_sigma_eps[i,:] = chain___kalman
    
```

Load sigma_eps chain for cpmmh

```python
R = 60000
burn_in = 10000

nbr_particles = str(N[0])

data_cpmmh_0999_sigma_eps = np.zeros((len(N),len(seeds), n))
data_cpmmh_099_sigma_eps = np.zeros((len(N),len(seeds), n))
```

```python
rho = "0.999"

for i in range(len(N)):
    for j in range(len(seeds)):

        nbr_particles = str(N[i])
        seed = str(seeds[j])

        job = seed + "_" + M_subjects + "_" + N_time + "_" + nbr_particles + "_" + rho

        chain_sigma_epsilon_cpmmh_tmp = np.loadtxt(open("data/SDEMEM OU/cpmmh for plot mess vs N/chain_sigma_epsilon_" + job + ".csv", "rb"), delimiter=",", skiprows=1)
        chain_sigma_epsilon_cpmmh_tmp = chain_sigma_epsilon_cpmmh_tmp[(burn_in+1):R]

        # select last 5k samples
        chain_sigma_epsilon_cpmmh_tmp = chain_sigma_epsilon_cpmmh_tmp[-n:]
        
        data_cpmmh_0999_sigma_eps[i,j,:] = chain_sigma_epsilon_cpmmh_tmp
```

```python
rho = "0.99"

for i in range(len(N)):
    for j in range(len(seeds)):

        nbr_particles = str(N[i])
        seed = str(seeds[j])

        job = seed + "_" + M_subjects + "_" + N_time + "_" + nbr_particles + "_" + rho

        chain_sigma_epsilon_cpmmh_tmp = np.loadtxt(open("data/SDEMEM OU/cpmmh for plot mess vs N/chain_sigma_epsilon_" + job + ".csv", "rb"), delimiter=",", skiprows=1)
        chain_sigma_epsilon_cpmmh_tmp = chain_sigma_epsilon_cpmmh_tmp[(burn_in+1):R]

        # select last 5k samples
        chain_sigma_epsilon_cpmmh_tmp = chain_sigma_epsilon_cpmmh_tmp[-n:]
        
        data_cpmmh_099_sigma_eps[i,j,:] = chain_sigma_epsilon_cpmmh_tmp
```

```python
data_cpmmh_099_sigma_eps.shape
```

Calc EM distance between chain___kalman and chain_sigma_epsilon_cpmmh_tmp

```python
a, b = np.ones((n,)) / n, np.ones((n,)) / n

data_emd_0999_sigma_eps = np.zeros((len(N),len(seeds)))
data_emd_099_sigma_eps = np.zeros((len(N),len(seeds)))
```

```python
data_runtimes_0999.shape
```

```python
data_runtimes_099
```

```python
for i in range(len(N)):
    for j in range(len(seeds)):

        M = ot.dist(data_kalman_sigma_eps[j,:].reshape((n, 1)), data_cpmmh_0999_sigma_eps[i,j,:].reshape((n, 1)))
        data_emd_0999_sigma_eps[i,j] = ot.emd(a, b, M, log=True)[1].get('cost')
```

```python
for i in range(len(N)):
    for j in range(len(seeds)):

        M = ot.dist(data_kalman_sigma_eps[j,:].reshape((n, 1)), data_cpmmh_099_sigma_eps[i,j,:].reshape((n, 1)))
        data_emd_099_sigma_eps[i,j] = ot.emd(a, b, M, log=True)[1].get('cost')
```
