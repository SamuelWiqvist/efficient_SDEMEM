## Compare PMMH, CPMMH, and block-wise CPMMH for the Ricker model

We will estimate the parameters of the Ricker model using differnet psudo-marginal algorithms. 

### Setput 

Set correct working directory. 

```julia
cd("..") # set correct working directory 
pwd() # this path should go to the folder with the .git file 
```

Load Ricker model and functions.  

```julia
using Random 
include(pwd()*"/src/Ricker model/rickermodel.jl"); 
```

Load data. 

```julia
y = CSV.read("src/Ricker model/y_data_set_2.csv"; allowmissing=:auto)[:,1]; 
```

Plot data.

```julia
PyPlot.figure()
PyPlot.plot(y)
PyPlot.xlabel("time")
PyPlot.ylabel("y");
```

### PMMH

Set up MCMC problem 

```julia
problem = set_up_problem(ploton = false)
problem.alg_param.N = 100
problem.alg_param.R = 10000
problem.alg_param.burn_in = 200
problem.data.y = y
problem.alg_param.print_interval = 5000 #problem.alg_param.R
problem.alg_param.alg = "pmmh"

# use fixed covaraince matrix and start at true parameter values to make comparisons easier 
problem.model_param.theta_0 = problem.model_param.theta_true
problem.adaptive_update = noAdaptation([0.0186481 -0.00436585 0.0177649; -0.00436585 0.00160763 -0.00397262; 0.0177649 -0.00397262 0.168024]);
```

Run MCMC algorithm. 

```julia
Random.seed!(1234)
@time res_PMMH = mcmc(problem); 

Theta = res_PMMH.Theta_est
loglik = res_PMMH.loglik_est
accept_vec = res_PMMH.accept_vec
prior_vec = res_PMMH.prior_vec;
```

Plotting

```julia
PyPlot.figure()
PyPlot.plot(Theta[1,:])
PyPlot.plot(problem.model_param.theta_true[1]*ones(size(Theta,2)), "k")
PyPlot.figure()
PyPlot.plot(Theta[2,:])
PyPlot.plot(problem.model_param.theta_true[2]*ones(size(Theta,2)), "k")
PyPlot.figure()
PyPlot.plot(Theta[3,:])
PyPlot.plot(problem.model_param.theta_true[3]*ones(size(Theta,2)), "k");
```

### CPMMH

```julia
problem.alg_param.alg = "cpmmh"
Random.seed!(1234)
@time res_CPMMH = cpmmh(problem, 0.99)

Theta = res_CPMMH.Theta_est
loglik = res_CPMMH.loglik_est
accept_vec = res_CPMMH.accept_vec
prior_vec = res_CPMMH.prior_vec;
```

Plotting

```julia
PyPlot.figure()
PyPlot.plot(Theta[1,:])
PyPlot.plot(problem.model_param.theta_true[1]*ones(size(Theta,2)), "k")
PyPlot.figure()
PyPlot.plot(Theta[2,:])
PyPlot.plot(problem.model_param.theta_true[2]*ones(size(Theta,2)), "k")
PyPlot.figure()
PyPlot.plot(Theta[3,:])
PyPlot.plot(problem.model_param.theta_true[3]*ones(size(Theta,2)), "k");
```

### Block-wise CPMMH

```julia
problem.alg_param.N = 10
problem.alg_param.alg = "blockwisecpmmh"
Random.seed!(1234)
@time res_blockwisecpmmh = blockwisecpmmh(problem, 10)

Theta = res_blockwisecpmmh.Theta_est
loglik = res_blockwisecpmmh.loglik_est
accept_vec = res_blockwisecpmmh.accept_vec
prior_vec = res_blockwisecpmmh.prior_vec;
```

```julia
PyPlot.figure()
PyPlot.plot(Theta[1,:])
PyPlot.plot(problem.model_param.theta_true[1]*ones(size(Theta,2)), "k")
PyPlot.figure()
PyPlot.plot(Theta[2,:])
PyPlot.plot(problem.model_param.theta_true[2]*ones(size(Theta,2)), "k")
PyPlot.figure()
PyPlot.plot(Theta[3,:])
PyPlot.plot(problem.model_param.theta_true[3]*ones(size(Theta,2)), "k");
```
