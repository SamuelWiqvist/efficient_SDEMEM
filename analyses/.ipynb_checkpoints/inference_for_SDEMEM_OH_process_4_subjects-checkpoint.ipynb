# Inference for the SDEMEM Ornstein-Uhlenbeck process 

We will estiamte the parameters for a SDEMEM Ornstein-Uhlenbeck process using different version of the Metroplis-Hastings within Gibbs algorithm. We will here cosider a SDEMEM model with 4 subjects. 

### The SDEMEM  Ornstein-Uhlenbeck process

The state model follows the OU process

$dX_t^{i} = \theta_1(\theta_2^{i}-X_t^{i})d_t + \theta_3dW_t, i = 1,2,..,M$. 

Where $M$ is the number of subjects, and the reversion levels $\theta_2^{i}$ are the random effects. The model for the random effects is 

$\theta_2^{i} \sim N(\mu_{\theta_2}, \sigma_{\theta_2})$.

The observation model is 

$Y_t^{i} = X_t^{i} + \epsilon_{t}^{i}, \epsilon_{t}^{i} \sim N(0, \sigma_{\epsilon})$.

Now, following the notation in [Picchini and Forman (2016)](https://arxiv.org/abs/1607.02633) we have that: $\kappa = [\theta_1, \theta_3]$, $\eta = [\mu_{\theta_2}, \sigma_{\theta_2}]$, $\phi^i = \theta_2^{i}$, and $\sigma_{\epsilon} = \sigma_{\epsilon}$.

Our goal is to sample from the full posterior

$\pi(\eta, \kappa, \sigma_{\epsilon}, \phi | y) \propto \pi(\eta)\pi(\kappa)\pi(\sigma_{\epsilon})\prod_{i = 1}^{M}\pi(\phi^{i})|\eta)\pi(y^{i}|\kappa, \sigma_{\epsilon},\phi^{i})$.

However, for this model is the likelihood $\pi(y^{i}|\kappa, \sigma_{\epsilon},\phi^{i})$ intractable and we will use variuse methods (Kalman filter, PMMH, CPMMH) to compute/estimate the likelihood.

### Metroplis-Hastings within Gibbs algorithm

We will use the Metroplis-Hastings within Gibbs algorithm to samlpe from the full posterior. The Metroplis-Hastings within Gibbs scheme follows:

1. Sample $\pi(\sigma_{\epsilon},\kappa |\phi, y ) \sim \pi(\sigma_{\epsilon})\pi(\kappa)\prod_{i = 1}^{M}\pi(y^{i}|\kappa, \sigma_{\epsilon},\phi^{i})$
2. For $i = 1,2,..,M$ sample $\pi(\phi^{i}|\sigma_{\epsilon},\kappa, \eta, y ) \sim \pi(\phi^{i}|\eta)\pi(y^{i}|\kappa, \sigma_{\epsilon},\phi^{i})$
3. Sample $\pi(\eta|\phi) \sim \pi(\eta)\prod_{i = 1}^{M}\pi(\phi^{i}|\eta)$
