# Inference for the SDEMEM Ornstein-Uhlenbeck process 

We will estiamte the parameters for a SDEMEM Ornstein-Uhlenbeck process using different version of the Metroplis-Hastings within Gibbs algorithm. We will here cosider a SDEMEM model with 1 subject. 

### Setput 

Set correct working directory. 

```julia
cd("..") # set correct working directory 
pwd() # the path should go to the folder with the .git file 
```

Load data and functions 

```julia
using Pkg
using PyPlot
using LinearAlgebra
using KernelDensity 
using Distributions 

# load functions
include(pwd()*"/src/SDEMEM OU process/ou_sdemem.jl")
include(pwd()*"/src/SDEMEM OU process/mcmc.jl")

N_time = 200
M_subjects = 1

#θ_1 = 0.5 # data w low obs noise 
#σ_ϵ = 0.1
#θ_3 = 0.05

θ_1 = 0.5 # data w high obs noise
σ_ϵ = 0.2
θ_3 = 0.1

y,x,t_vec,dt,η,κ,σ_ϵ,ϕ,prior_η,prior_κ,prior_σ_ϵ = set_up(σ_ϵ = σ_ϵ, θ_3 = θ_3,θ_1=θ_1, M=M_subjects,N=N_time)
```

Plot latent process (left) and observed process (rigth).

```julia
PyPlot.figure(figsize=(8,5))
PyPlot.subplot(121)
PyPlot.plot(t_vec,Array(x')); 
PyPlot.subplot(122)
PyPlot.plot(t_vec,Array(y'));
```

Set parameters for MH within Gibbs algorithm.

```julia
R = 10000
burn_in = 2000

cov_1 = Array(Diagonal([0.001;0.001;0.001]))
cov_2 = Array(Diagonal(0.2*ones(M_subjects)))
cov_3 = Array(Diagonal([0.8;0.8]))

startval_1 = [1;1;1] #[0.2;0.8;0.1] # we set the start values to the true values just ot test the algorithm
startval_2 = 2*ϕ
startval_3 = [10;10]; # [4.0;0.8]
```

Setting for trace plots.

```julia
plot_from_start = false # set to true to plot chains from first iteration
burn_in_trace = 1

if !plot_from_start
    burn_in_trace = burn_in
end 
```

### Load results  Kalman filter 

Runtime: 2.3 sec. 

```julia
run_time_kalman = 2.3 

job = string(M_subjects)*"_"*string(N_time)

chain_1_kalman = Matrix(CSV.read("data/SDEMEM OU/kalman/chain_1_"*job*".csv"; allowmissing=:auto))'
chain_2_kalman = Matrix(CSV.read("data/SDEMEM OU/kalman/chain_2_"*job*".csv"; allowmissing=:auto))'
chain_3_kalman = Matrix(CSV.read("data/SDEMEM OU/kalman/chain_3_"*job*".csv"; allowmissing=:auto))'
        
accept_vec_kalman = Matrix(CSV.read("data/SDEMEM OU/kalman/accept_vec_"*job*".csv"; allowmissing=:auto))'
```

### Plot posterior dist 

Posterior for $\sigma_{\epsilon}$. Blue line kalman, red line PMMH, dashed line CPMMH, dotted line block-CPMMH, green line prior, black verticla line true paramter value.

```julia
prior_grid = LinRange(0,1,100)
prior = pdf.(Gamma(prior_σ_ϵ[1].parameters[1],prior_σ_ϵ[1].parameters[2]), prior_grid)
posterior_kalman = kde(chain_1_kalman[1,burn_in:end])
posterior_pmmh = kde(chain_1_pmmh[1,burn_in:end])
posterior_cpmmh = kde(chain_1_cpmmh[1,burn_in:end])
#posterior_bcpmmh = kde(chain_1_bcpmmh[1,burn_in:end])

PyPlot.figure()
PyPlot.plot(prior_grid,prior, "g")
PyPlot.plot(posterior_kalman.x,posterior_kalman.density, "b")
PyPlot.plot(posterior_pmmh.x,posterior_pmmh.density, "r")
PyPlot.plot(posterior_cpmmh.x,posterior_cpmmh.density, "--r")
#PyPlot.plot(posterior_bcpmmh.x,posterior_bcpmmh.density, ":r")
PyPlot.plot((σ_ϵ, σ_ϵ), (0, maximum(posterior_kalman.density)), "k")
PyPlot.xlim((minimum(posterior_kalman.x)-0.02, maximum(posterior_kalman.x)+0.02));
```
